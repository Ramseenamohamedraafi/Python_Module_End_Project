# Providing a dataset of employees working in ABC company. It consists of 458 rows and 9 columns. The company needs the detailed report and explanation of their employees in each team.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_details=pd.read_csv('myexcel - myexcel.csv.csv')
employee_details
```

```python
employee_details['Height']=np.random.randint(150,180,size=len(employee_details))
```

# 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
number=employee_details.Team.value_counts()
percentage_split=(number/len(employee_details))*100
Team=pd.DataFrame({'Team':number.index,'counts':number.values,'percentage':percentage_split.values})
Team
```

# 2.Segregate the employees w.r.t different positions.

```python
Position_count=employee_details.groupby('Position')
print(Position_count.value_counts())
```

# 3.Find from which age group most of the employees belong to.

```python
age_grp=pd.cut(employee_details['Age'],bins=[18,25,35,45,55,65])
age_grp_counts=age_grp.value_counts()
print(age_grp_counts) 
```

# 4.Find out under which team and position, spending in terms of salary is high.

```python
grp_data=employee_details.groupby(['Team','Position'])['Salary'].sum()
max_spending=grp_data.idxmax()
team_with_highest_spending,position_with_highest_spending=max_spending
print("Team with highest spending:",team_with_highest_spending)
print("Position with highest spending:",position_with_highest_spending)
```

# 5.Find if there is any correlation between age and salary , represent it visually.

```python
employee_details.plot.scatter(x='Age',y='Salary')
plt.title('correlation between AGE and SALARY')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
